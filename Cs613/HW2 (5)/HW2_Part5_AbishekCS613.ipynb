```python

def entropy(target_col):#y_train
   
    elements,counts = np.unique(target_col,return_counts = True)
    #print(elements.shape)
    #print(counts)
    entropy = np.sum([(-counts[i]/np.sum(counts))*np.log2(counts[i]/np.sum(counts)) for i in range(len(elements))])
    return entropy
##log2 and log make a difference in entropy calculation
entropy(y_train)
```

```python
def InfoGain(data,split_attribute_,target_train):
 
    
    #Calculate the entropy of the total dataset

    total_entropy = entropy([target_train]) 
    #print(total_entropy)
    
    ##Calculate the entropy of the dataset
    Vals=np.zeros(2)
    #print(Vals)
    #Calculate the values and the corresponding counts for the split attribute 
    vals,counts= np.unique(split_attribute_,return_counts=True)
    #print("vals:", vals, "counts: ", counts)
    Vals = vals
    #print("Vals:", Vals, "counts: ", counts)
    #for i in range(len(Vals)):
          #  if Vals[i]:
                #print("Found element")
            #else:
                #print(len(Vals))
              #  Vals[i] == 0
            
    
    target_col = []
    target_col_EFones = []
    target_col_EFzeros = []
    #print(split_attribute_.shape[0])
    for j in range(len(split_attribute_)):
        for i in range(len(Vals)):  
            if (Vals[i] == 1 and split_attribute_[j]==Vals[i]):#Ef1
                #print(j)
                #print(vals[i])
                #print(split_attribute_[j])
                #print(target_train[j])
                target_col_EFones.append(target_train[j])
                #print("ones")
            elif (Vals[i] == 0 and split_attribute_[j]==Vals[i]):#Ef0
                #print(j)
                #print(split_attribute_[j,0])
                #print(target_train[j])
                target_col_EFzeros.append(target_train[j])
                #print("zeros")
                #print("y-train value to be appended for entropy target column: ", target_train[j], "and index: ", j)
    
    #print(len(target_col_EFones))
    target_col_EFones = np.reshape(np.asarray(target_col_EFones), (np.asarray(target_col_EFones).shape[0], 1))
    target_col_EFzeros = np.reshape(np.asarray(target_col_EFzeros), (np.asarray(target_col_EFzeros).shape[0], 1))
    #print(target_col_EFones) 
    #print(target_col_EFzeros)            
    
    #Calculate the weighted entropy
    Weighted_Entropy_ones =np.asarray([(counts[0]/np.sum(counts))*entropy(target_col_EFones) ])
    
    
    
    if len(counts)<2:
        
        Weighted_Entropy_zeros =np.asarray([(0/np.sum(counts))*entropy(target_col_EFzeros) ])
    else:
        Weighted_Entropy_zeros =np.asarray([(counts[1]/np.sum(counts))*entropy(target_col_EFzeros) ])
    #print(len(counts))   
    #print(Weighted_Entropy_ones)
    #print(Weighted_Entropy_zeros)
    Weighted_Sum = Weighted_Entropy_ones + Weighted_Entropy_zeros
    #print(Weighted_Sum)
    #Calculate the information gain
    Information_Gain = total_entropy - Weighted_Sum
    print(Information_Gain)
    return Information_Gain

temp = X_train_std_bin[:,0:1]
Y_train_bin = X_train_std_bin[:, -1:]
#print(temp)
InfoGain(X_train_std_bin, temp, Y_train_bin)
```

```python
def ID3(data,originaldata,parent_node_class = None):

    features = {}
    if data.shape[0]>1:
        print("sub data shape: ", data.shape)
        for i in range(data.shape[1]-1):
    #print(X_train_std_bin[:, i].shape)
            features.update( {i : data[:, i]} )
    
    #Define the stopping criteria --> If one of this is satisfied, we want to return a leaf node#
    
    
    #If all target_values have the same value, return this value
    target_train = data[:, -1:]
    #print("feature length: ", len(features))
    
    if len(np.unique(target_train)) <= 1:
        return np.unique(target_train)[0]

    
    #If the dataset is empty, return the mode target feature value in the original dataset
    elif data.shape[0] <= 1:
    #elif len(data) == 0:
        return np.unique(target_train)[np.argmax(np.unique(target_train,return_counts=True)[1])]
        
    
    
    ##determine features of sub_dataset here
    
    
    elif len(features) == 0:
        print(features)
        return parent_node_class
    
    
    
    else:
        # default value for this node --> The mode target feature valu
        parent_node_class = np.unique(target_train)[np.argmax(np.unique(target_train,return_counts=True)[1])]
        
        #Select the feature which best splits the dataset
        
        
        IG_values = [InfoGain(data,features[feature],target_train) for feature in features] #Return the information gain values for the features in the dataset
            
        best_feature_index = np.argmax(IG_values)
        #print("best feature index: ", best_feature_index)
        #print(features[best_feature_index])
        best_feature = features[best_feature_index]
        #print(best_feature)
        
        #Create the tree structure. The root gets the name of the feature (best_feature) with the maximum information
        #gain in the first run
        tree = {best_feature_index:{}}
        
        
        #Remove the feature with the best inforamtion gain from the feature space
        Features = [features[i] for i in features if i != best_feature_index]
        Features = (np.asarray(Features)).T
        #print("Features shape: ", Features.shape)
        
        #Grow a branch under the root node for each possible value of the root node feature
        
        for value in np.unique(best_feature):
            value = value
            #print(value)
            #Split the dataset along the value of the feature with the largest information gain and therwith create sub_datasets
            print("New features number of columns: ", Features.shape)
            if Features.shape[0]==0:
                print("Features are empty")
                sub_data = np.empty((1, 1), dtype=float)
               
            else:
                sub_data = np.empty((0, Features.shape[1]+1), dtype=float)
            #sub_data = []
            
            
            for i in range(len(best_feature)):
                if best_feature[i] == value:
                    #print("Feature i: ", Features[i].shape, "Features: ", Features[i])
                    #print("Target i: " , target_train[i])
                    if Features.shape[0]==0:
                        temp_data = target_train[i]
                    else:
                        temp_data = np.append(Features[i], (target_train[i]))
                    #print("temp data shape: ", temp_data.shape, "temp data for stack in sub data: ", temp_data)
                    #print("sub data shape: ", sub_data.shape)
                    sub_data = np.vstack((sub_data, temp_data))
              
            #print("Sub data: ", sub_data, "Sub data shape: ", sub_data.shape)
            #Call the ID3 algorithm for each of those sub_datasets with the new parameters --> Here the recursion comes in!
            subtree = ID3(sub_data,X_train_std_bin,parent_node_class)
            
            #Add the sub tree, grown from the sub_dataset to the tree under the root node
            print("best feature index: ", best_feature_index, "binary value:", value)
            tree[best_feature_index][value] = subtree
            
        return(tree)    
                
```

```python
def predict(query,tree,default = 1):
    for key in list(query.keys()):
        if key in list(tree.keys()):
            #2.
            try:
                result = tree[key][query[key]] 
            except:
                return default
  
            #3.
            result = tree[key][query[key]]
            #4.
            if isinstance(result,dict):
                return predict(query,result)

            else:
                return result
```
