```python
import json
import ast
from IPython.display import Image
import openpyxl
```

```python
with open('pose_positions.json') as main_json_file:
    data = main_json_file.readlines()
```

```python
data_con = json.loads(data[0])
```

```python
image_data = []
image_labels = []
image_name=[]
for c in range(len(data_con['annolist'])):
    if data_con['act'][c]['cat_name']!="" and data_con['annolist'][c]['annorect']!=[]:
        if type(data_con['annolist'][c]['annorect'])==list and 'annopoints' in data_con['annolist'][c]['annorect']:
            if data_con['annolist'][c]['annorect'][0]['annopoints']!=[]:
                image_labels.append(data_con['act'][c])
                image_data.append(data_con['annolist'][c]['annorect'][0]['annopoints']['point'])
                image_name.append(data_con['annolist'][c]['image']['name'])
        else:
            if 'annopoints' in data_con['annolist'][c]['annorect']:
                image_labels.append(data_con['act'][c])
                image_data.append(data_con['annolist'][c]['annorect']['annopoints']['point'])
                image_name.append(data_con['annolist'][c]['image']['name'])
```

```python
def joint_create(filename):
    wb = openpyxl.Workbook() 
    sheet = wb.active
    joint_data = []
    counter=1
    for x in range(len(image_data)):
        for y in image_data[x]:
            if y['id']==0:
                c1 = sheet.cell(row = counter, column = 1)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 2)
                c2.value = y['y']
            
            elif y['id']==1:
                c1 = sheet.cell(row = counter, column = 3)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 4)
                c2.value = y['y']
            
            elif y['id']==2:
                c1 = sheet.cell(row = counter, column = 5)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 6)
                c2.value = y['y']
            
            elif y['id']==3:
                c1 = sheet.cell(row = counter, column = 7)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 8)
                c2.value = y['y']
            
            elif y['id']==4:
                c1 = sheet.cell(row = counter, column = 9)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 10)
                c2.value = y['y']
            
            elif y['id']==5:
                c1 = sheet.cell(row = counter, column = 11)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 12)
                c2.value = y['y']
            
            elif y['id']==6:
                c1 = sheet.cell(row = counter, column = 13)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 14)
                c2.value = y['y']
            
            elif y['id']==7:
                c1 = sheet.cell(row = counter, column = 15)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 16)
                c2.value = y['y']
            
            elif y['id']==8:
                c1 = sheet.cell(row = counter, column = 17)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 18)
                c2.value = y['y']
            
            elif y['id']==9:
                c1 = sheet.cell(row = counter, column = 19)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 20)
                c2.value = y['y']
            
            elif y['id']==10:
                c1 = sheet.cell(row = counter, column = 21)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 22)
                c2.value = y['y']
            
            elif y['id']==11:
                c1 = sheet.cell(row = counter, column = 23)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 24)
                c2.value = y['y']
            
            elif y['id']==12:
                c1 = sheet.cell(row = counter, column = 25)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 26)
                c2.value = y['y']
                
            elif y['id']==13:
                c1 = sheet.cell(row = counter, column = 27)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 28)
                c2.value = y['y']
            
            elif y['id']==14:
                c1 = sheet.cell(row = counter, column = 29)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 30)
                c2.value = y['y']
            
            elif y['id']==15:
                c1 = sheet.cell(row = counter, column = 31)
                c1.value = y['x']
                c2 = sheet.cell(row = counter, column = 32)
                c2.value = y['y']
        
        c3 = sheet.cell(row = counter, column = 33)
        c3.value = image_labels[x]['cat_name']
        c4 = sheet.cell(row = counter, column = 34)
        c4.value = image_name[x]
        counter+=1
    wb.save(filename)
```
